```python
import time as t
from ctm_python_client.core.workflow import Workflow, WorkflowDefaults
from ctm_python_client.core.comm import Environment
from aapi import *

workflow = Workflow(Environment.create_workbench('workbench'), WorkflowDefaults(run_as='workbench'))

workflow.clear_all()
job = JobCommand('JobEvent', command='ls -lt')

job.event_list.append(AddEvents([EventOutAdd(event="e2",date=Event.Date.NoDate)]))
job.event_list.append(AddEvents([EventOutAdd(event="e3",date=Event.Date.NoDate)]))

job.event_list.append(DeleteEvents([Event(event="e5",date=Event.Date.NextOrderDate)]))
job.event_list.append(DeleteEvents([Event(event="e4",date=Event.Date.NextOrderDate)]))

job.event_list.append(WaitForEvents(["(",Event(event="e1"),"AND",Event(event="e2"), ")","OR", Event(event="e3", date=Event.Date.AnyDate)]))
job.event_list.append(WaitForEvents(["(",Event(event="e2"),"OR",Event(event="e4"), ")"]))

print(job.dumps_aapi(indent=2))
workflow.add(job, inpath='EventsFolder')

run = workflow.run()
```
